### Breeden-Lintzenberger

```python
expiry = '2022-01-07'
calls_expiry = calls[calls['expiry']==expiry]
calls_expiry.loc[:, 'short_asof_date'] = pd.to_datetime(calls_expiry.loc[:,'short_asof_date'])
```

```python
filename = 'DTB3.xls'

path = os.getcwd() + '\\' + filename

rf = pd.read_excel(filename)
rf['DTB3'] = rf['DTB3']/100
rf = rf.rename(columns={'DTB3':'rfr'})
```

```python
calls_expiry.dtypes
```

```python
calls_expiry = pd.merge(left=calls_expiry, right=rf, how='left', left_on='short_asof_date', right_on='observation_date')
```

```python
def BSPrice(vol, S, K, r, T, opt_type):
    
    d1 = (np.log(S/K) + (r + 0.5* vol)*T)/(vol * np.sqrt(T))
    d2 = d1 - vol * np.sqrt(T)
    
    if 'c' in opt_type.lower():
        BS = S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)
    else:
        BS = K * np.exp(-r * T) * norm.cdf(-d2) - S * norm.cdf(-d1)
        
    return BS
```

```python
def BSPriceVsTheor(vol, S, K, r, T, opt_type, P):
    
    BS = BSPrice(vol, S, K, r, T, opt_type)
    return (BS - P)**2
```

```python
expiry_in_days = calls_expiry['expiry'] - calls_expiry['asof_date']
calls_expiry['expiryFrac'] = expiry_in_days.apply(lambda x: x.total_seconds() / (60*60*24*256))
```

```python
calls_expiry.iloc[20,:]
```

```python
BSPrice(0.1146, calls_expiry['last_px_underlying'].iloc[20], calls_expiry['strike'].iloc[20], 0.0055, calls_expiry['expiryFrac'].iloc[20], calls_expiry['type'].iloc[20])
```

```python
def IVFinder(P, S, K, r, T, opt_type):
    
    return opt.minimize(BSPriceVsTheor, x0=0.05, args=(S,K,r,T,opt_type,P))['x'][0]
```

```python
P = calls_expiry['last'].iloc[20]
S = calls_expiry['last_px_underlying'].iloc[20]
K = calls_expiry['strike'].iloc[20]
r = 0.0055
T = calls_expiry['expiryFrac'].iloc[20]
opt_type = calls_expiry['type'].iloc[20]

```

```python
calls_expiry['iv2'] = calls_expiry.apply(lambda row: IVFinder(row['last'], row['last_px_underlying'], row['strike'], 0.0055, row['expiryFrac'], row['type']), axis=1)
```

#### Gaussian kernel smoothing

```python
x = np.linspace(calls_expiry['strike'].iloc[0] ,calls_expiry['strike'].iloc[-1],calls_expiry.shape[0])
y = calls_expiry['iv']
# convert both to arrays
x_sm = np.array(x)
y_sm = np.array(y)

# resample to lots more points - needed for the smoothed curves
x_smooth = np.linspace(x_sm.min(), x_sm.max(), 200)

# spline - always goes through all the data points x/y
# y_spline = interpolate.spline(x, y, x_smooth)

spl = interpolate.UnivariateSpline(x, y)

sigma = 2
x_g1d = ndimage.gaussian_filter1d(x_sm, sigma)
y_g1d = ndimage.gaussian_filter1d(y_sm, sigma)
```

```python
sns.lineplot(x=x_g1d, y=y_g1d)
sns.lineplot(x=x, y=y)
plt.show()
```

#### Using the smoothed curve to compute the PDF

```python
strikes = x_g1d
prices = np.zeros(len(strikes))

for i, K in enumerate(strikes):
    prices[i] = BSPrice(y_g1d[i], calls_expiry['last_px_underlying'].iloc[0], strikes[i], 0.0055, calls_expiry['expiryFrac'].iloc[0], calls_expiry['type'].iloc[0])
    
```
